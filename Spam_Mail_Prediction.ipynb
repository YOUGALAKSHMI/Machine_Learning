import dependicies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and preprocessing

```python
data=pd.read_csv('/content/mail_data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
#replacing null values with empty string
mail_data=data.where((pd.notnull(data)),'')
```

```python
mail_data.head()
```

```python
#encode label into numerical value
#spam mail : 0 ham :1
mail_data.loc[mail_data['Category'] == 'spam' ,'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham' ,'Category',] = 1
```

```python
mail_data.head()
```

```python
#train test split
x=mail_data['Message']
y=mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

Feature extraction

```python
#coverting message into meaningful numerical value
feature_extraction=TfidfVectorizer(min_df=1, stop_words='english',  lowercase=True)

x_train_features =feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)

#convert y_train and y_test to integers

y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
x_train_features.shape
```

```python
x_test_features.shape
```

```python
print(x_train_features)
```

```python
print(x_test_features)
```

Training Model

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

Model Evaluation

```python
# prediction on train data
prediction_on_training_data=model.predict(x_train_features)
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data)
```

```python
print(accuracy_on_training_data)
```

```python
# prediction on test data
prediction_on_test_data = model.predict(x_test_features)
accuracy_on_testing_data = accuracy_score(y_test, prediction_on_test_data)
```

```python
print(accuracy_on_testing_data)
```

Predicit using new message

```python
#build a predective system
input_mail=["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]

#convert text to feature extract
input_feature=feature_extraction.transform(input_mail)

#make prediction
prediction=model.predict(input_feature)
print(prediction)

if prediction[0]==1:
   print("ham mail")
else:
  print("spam mail")
```
